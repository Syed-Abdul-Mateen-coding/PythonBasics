#**Question 6)**
* Generators are a simple way to create iterators using functions and the yield keyword instead of returning values.

* yield returns a value and pauses the function’s state, allowing it to resume from where it left off.

* Generators produce values one at a time and only when required, avoiding the need to generate or store all values at once, this enables efficient utilisation of memory.



#**Question 7)**

* Generators yield one item at a time, so they don't store the entire result in memory enabling efficient utilization of memory.


```
def count_up_to(n):
    for i in range(n):
        yield i
```

* Generators compute values only when needed, which improves performance, this is called as lazy evaluation.They pause after yield and resume from the same point when called again.


* we can build infinite loops safely using generators because they don’t try to store everything at once.

```
def infinite():
    i = 0
    while True:
        yield i
        i += 1

```
* With generators, there's no need to manually implement iter() and next(), just use yield.

#**Question 9)**
The map() function applies a given function to each item in an iterable (like a list, tuple, etc.) and returns a map object (an iterator) with the results.

syntax : map(function_to _be_applies,iterable)
* function can be lambda function also
* iterable can be list, tuple, set etc...



```
fruits = ["apple", "banana", "mango"]
result = map(str.upper, fruits)

print(list(result))  # Output: ['APPLES', 'BANANAS', 'MANGOS']
```

````python
#7)
def read_file(file):
  with open(file,"r") as f:
    for line in f:
      yield line

obj=read_file("file.txt")
next(obj)
````
